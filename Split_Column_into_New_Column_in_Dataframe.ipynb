```python
import pandas as pd
```

###Create Dataframe

```python
data = {'id':[1,2,3,4,5,6],
        'name':['John', 'Andy', 'Tom', 'Jimmy', 'Laura', 'Bob'],
        'mail':['john_123@yahoo.com',
                'Andy.321@gmail.com',
                'tom.909@hotmail.com',
                'JIMMY@HOTMAIL.COM',
                'laura@gmail.com',
                'BOB_999@GMAIL.COM']}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

###Create New Column by Mail Type

```python
email_type = ['yahoo', 'gmail', 'hotmail']
```

```python
pattern = '|'.join(email_type)

df.insert(3,'emailType', df['mail'].str.lower().str.extract( '('+pattern+')', expand=False))
```

```python
df.head()
```

###Create New Column as Username

For example we want to create a username column by taking their mail name. 

```python
df.insert(3, 'username', df['mail'].str.lower().str.split('@').str[0])
```

```python
df.head()
```

If we don't want to use special character such as "." and "_", we can remove it like this:

```python
df['username_new'] = df['username'].str.replace('.', '').str.replace('_','')
```
